```python
df = pd.read_csv("dataSet.csv")

# take a look at the dataset
df.head()
```

Lets select some features that we want to use for regression.


```python
cdf = df[['sqrt',	'washRoom',	'bedRoom',	'floor',	'lift',	'roodSize','areaIndex','SalePrice']]
cdf.tail(9)
```

```python
msk = np.random.rand(len(df)) < 1
train = cdf[msk]
test = cdf[~msk]
```

```python
print (train)
```

```python
from sklearn import linear_model
regr = linear_model.LinearRegression()
x = np.asanyarray(train[['sqrt',	'washRoom',	'bedRoom',	'floor',	'lift',	'roodSize','areaIndex']]) # here is the difference from simple regression
y = np.asanyarray(train[['SalePrice']])
regr.fit (x, y)
# The coefficients
print ('Coefficients: ', regr.coef_)
```

```python
import joblib
```

```python
filename= 'Finalized_model.sav'
joblib.dump(regr,filename)
```
